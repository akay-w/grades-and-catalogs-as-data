```python
import re
import string
import pandas as pd
import nltk
```

```python
#Changed this option so that I could see the full descriptions
#Otherwise they're abbreviated, since they're so long
pd.set_option('display.max_colwidth', None)
```

```python
#read in course data from Excel file
path = r"C:\Users\Mandy\Documents\Data science projects\Furman catalog\courses and descriptions, 20210510.xlsx"
df = pd.read_excel(path)
```

```python
df.head()
```

```python
#Check to see if there are any NA values
df.isnull().any()
```

```python
#Drop any row with NA values
df = df.dropna()
```

```python
#Add a new column to keep the original description for comparison purposes
df["Original Description"] = df["Description"]
df.head()
```

```python
#Use a regex to replace lines like "4 credits."
pattern = r'\d credits.'
df["Description"].replace(pattern, '', regex=True, inplace=True)
```

```python
#Check an affected row to make sure it worked
df.head(1)
```

```python
#Use regex to remove lines like Same as (a different class)
pattern2 = r"Same as [^.]*.\s"
df["Description"].replace(pattern2, '', regex=True, inplace=True)
```

```python
#Check an affected row to make sure it worked
df.loc[df['Course'] == "EES 343"]
```

```python
#Use regex to remove May Experience ONLY lines
pattern3 = r"May Experience ONLY."
df["Description"].replace(pattern3, '', regex=True, inplace=True)
```

```python
#Check an affected row to make sure it worked
df.loc[df['Course'] == "EES 357"]
```

```python
#Use regex to remove lines like A student cannot recieve credit for both this and this other course
pattern4 = r"A student cannot receive credit [\w\s\d,()-]+[.]"
df["Description"].replace(pattern4, '', regex=True, inplace=True)
```

```python
#Check an affected row to make sure it worked
df.loc[df['Name'] == "Introduction to Statistics"]
```

```python
#Check an affected row to make sure it worked
df.loc[df['Name'] == "Finite Mathematics"]
```

```python
#Use regex to remove any remaining course names of format ECN 200, ECN-200, and ECN200
pattern5 = "[A-Z][A-Z][A-Z](\s)?-?\d+"
df["Description"].replace(pattern5, '', regex=True, inplace=True)
```

```python
#Check an affected row to make sure it worked
df.loc[df['Course'] == "HUM 102"]
```

```python
#Use regex to remove any remaining words with numbers
#Check affected rows to make sure it worked
pattern6 = r"\d+"
df["Description"].replace(pattern6, '', regex=True, inplace=True)
df.head(11)
df.loc[df['Course'] == "ANT 401"]
```

```python
#Check to make sure decimals were removed
df.loc[df['Course'] == "MXP 220"]
```

```python
#Replace all dashes with spaces
df['Description'] = df['Description'].str.replace("-", " ")
#Check affected rows to make sure it worked
df.loc[df["Name"] == "Art for Non-Art Majors"]
```

```python
#Change everything to lowercase
df["Description"] = df["Description"].str.lower()
```

```python
#Use nltk to tokenize the text
df["Description"] = df["Description"].apply(nltk.word_tokenize)
df.head()
```

```python
#remove punctuation
df["Description"] = df["Description"].apply(lambda x: [word for word in x if word not in string.punctuation])
df.head()
```

```python
#remove other weird characters
to_remove = ['``', '"', "''"]
df["Description"] = df["Description"].apply(lambda x: [word for word in x if word not in to_remove])
df.loc[df["Course"] == "ANT 302"]
```

```python
#remove any stopwords
df["Description"] = df["Description"].apply(lambda x: [word for word in x if word not in nltk.corpus.stopwords.words("english")])
df.head()
```

```python
#Remove any single quotes lingering at the beginning of tokens
def remove_leading_quote(data):
    new_data = []
    for token in data:
        if token[0] == "'":
            new_data.append(token[1:])
        else:
            new_data.append(token)
    return new_data
df["Description"] = df["Description"].apply(remove_leading_quote)
df.loc[df["Course"] == "FYW 1272"]
```

```python
#Remove any tokens smaller than 3 characters
df["Description"] = df["Description"].apply(lambda x: [word for word in x if len(word) > 2])
df.loc[df["Course"] == "FYW 1272"]
```

```python
#download nltk wordnet to use lemmatizer if you don't have it
#nltk.download('wordnet')
```

```python
#use nltk lemmatize to lemmatize tokens
lemmatizer = nltk.WordNetLemmatizer()
df["Description"] = df["Description"].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])
df.head()
```

```python
#scroll through a bit to try to catch anything weird left behind
df[50:100]
```

```python
#remove remaining weird characters and unimportant words
to_remove = ['and/or', 'topic', 'include', "student", "study", "course", "including", "ger"]
df["Description"] = df["Description"].apply(lambda x: [word for word in x if word not in to_remove])
df.head(50)
```
